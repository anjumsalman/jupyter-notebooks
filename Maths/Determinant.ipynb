## Representation
**Case 1:** Consider the equations:
$$a_1x + b_1 = 0$$
$$a_2x + b_2 = 0$$

Eliminating $x$, we get:
$$a_1b_2 - a_2b_1 = 0$$
Which as determinant can be written as:
$$\begin{vmatrix} 
a_1 & b_1 \\ 
a_2 & b_2 
\end{vmatrix} = 0$$

**Case 2:** Now consider two linear equations:
$$a_1x + b_1y + c_1 = 0$$
$$a_2x + b_2y + c_2 = 0$$
To get $x$, we eliminate $y$ to get:
$$c_1b_2 + a_1b_2x = b_1c_2 + a_2b_1x$$
$$\Rightarrow x(a_1b_2 - a_2b_1) + (c_1b_2 - b_1c_2) = 0$$
$$\Rightarrow x\begin{vmatrix} 
a_1 & b_1 \\ 
a_2 & b_2 
\end{vmatrix} + \begin{vmatrix} 
c_1 & b_1 \\ 
c_2 & b_2 
\end{vmatrix} = 0$$
$$\Rightarrow x = -\frac{\begin{vmatrix} 
c_1 & b_1 \\ 
c_2 & b_2 
\end{vmatrix}}{\begin{vmatrix} 
a_1 & b_1 \\ 
a_2 & b_2 
\end{vmatrix}}$$
$$\Rightarrow x = \frac{\begin{vmatrix} 
b_1 & c_1 \\ 
b_2 & c_2 
\end{vmatrix}}{\begin{vmatrix} 
a_1 & b_1 \\ 
a_2 & b_2 
\end{vmatrix}}$$
Similarly, $y$ is:
$$y = \frac{\begin{vmatrix} 
c_1 & a_1 \\ 
c_2 & a_2 
\end{vmatrix}}{\begin{vmatrix} 
a_1 & b_1 \\ 
a_2 & b_2 
\end{vmatrix}}$$

**Case 3:** What about 3 equations with two unknowns:
$$a_1x + b_1y + c_1 = 0 \ \ \ \ \ (1)$$
$$a_2x + b_2y + c_2 = 0 \ \ \ \ \ (2)$$
$$a_3x + b_3y + c_3 = 0 \ \ \ \ \ (3)$$

Solving $2$ and $3$, we get:
$$x = \frac{\begin{vmatrix} 
b_2 & c_2 \\ 
b_3 & c_3 
\end{vmatrix}}{\begin{vmatrix} 
a_2 & b_2 \\ 
a_3 & b_3 
\end{vmatrix}}$$
$$y = \frac{\begin{vmatrix} 
c_2 & a_2 \\ 
c_3 & a_3 
\end{vmatrix}}{\begin{vmatrix} 
a_2 & b_2 \\ 
a_3 & b_3 
\end{vmatrix}}$$

Putting this in $1$, we get:
$$a_1\frac{\begin{vmatrix} 
b_2 & c_2 \\ 
b_3 & c_3 
\end{vmatrix}}{\begin{vmatrix} 
a_2 & b_2 \\ 
a_3 & b_3 
\end{vmatrix}} + b_1\frac{\begin{vmatrix} 
c_2 & a_2 \\ 
c_3 & a_3 
\end{vmatrix}}{\begin{vmatrix} 
a_2 & b_2 \\ 
a_3 & b_3 
\end{vmatrix}} + c_1 = 0$$
$$\Rightarrow a_1\begin{vmatrix} 
b_2 & c_2 \\ 
b_3 & c_3 
\end{vmatrix} + b_1\begin{vmatrix} 
c_2 & a_2 \\ 
c_3 & a_3 
\end{vmatrix} + c_1\begin{vmatrix} 
a_2 & b_2 \\ 
a_3 & b_3 
\end{vmatrix} = 0$$
$$\Rightarrow a_1\begin{vmatrix} 
b_2 & c_2 \\ 
b_3 & c_3 
\end{vmatrix} - b_1\begin{vmatrix} 
a_2 & c_2 \\ 
a_3 & c_3 
\end{vmatrix} + c_1\begin{vmatrix} 
a_2 & b_2 \\ 
a_3 & b_3 
\end{vmatrix} = 0$$

Which in terms of 3x3 determinant is:
$$\begin{vmatrix} 
a_1 & b_1 & c_1 \\ 
a_2 & b_2 & c_2 \\
a_3 & b_3 & c_3
\end{vmatrix} = 0$$

The above determinant basically represents the condition for the 3 linear equations to have common solution. Above determinant is the relation below
$$a_1(b_2c_3 - c_2b_3) - b_1(a_2c_3 - a_3c_2) + c_1(a_2b_3 - a_3b_2) = 0$$

**Case 4:** Finally, consider 3 equations with 3 unknowns:
$$a_1x + b_1y + c_1z = d_1 \ \ \ \ \ (1)$$
$$a_2x + b_2y + c_2z = d_2 \ \ \ \ \ (2)$$
$$a_3x + b_3y + c_3z = d_3 \ \ \ \ \ (3)$$

To get $x$, we eliminate $y$ and $z$ to get:
$$\begin{vmatrix} 
a_1x - d_1 & b_1 & c_1 \\ 
a_2x - d_2 & b_2 & c_2 \\
a_3x - d3 & b_3 & c_3
\end{vmatrix} = 0$$
$$\Rightarrow x\begin{vmatrix} 
a_1 & b_1 & c_1 \\ 
a_2 & b_2 & c_2 \\
a_3 & b_3 & c_3
\end{vmatrix} - \begin{vmatrix} 
d_1 & b_1 & c_1 \\ 
d_2 & b_2 & c_2 \\
d_3 & b_3 & c_3
\end{vmatrix} = 0$$

We get the following equations:
$$x = \frac{{\Delta}_1}{\Delta}, y = \frac{{\Delta}_2}{\Delta}, z = \frac{{\Delta}_3}{\Delta}$$
Where
$$\Delta = \begin{vmatrix} 
a_1 & b_1 & c_1 \\ 
a_2 & b_2 & c_2 \\
a_3 & b_3 & c_3
\end{vmatrix}$$
$${\Delta}_1 = \begin{vmatrix} 
d_1 & b_1 & c_1 \\ 
d_2 & b_2 & c_2 \\
d_3 & b_3 & c_3
\end{vmatrix}$$
$${\Delta}_2 = \begin{vmatrix} 
a_1 & d_1 & c_1 \\ 
a_2 & d_2 & c_2 \\
a_3 & d_3 & c_3
\end{vmatrix}$$
$${\Delta}_3 = \begin{vmatrix} 
a_1 & b_1 & d_1 \\ 
a_2 & b_2 & d_2 \\
a_3 & b_3 & d_3
\end{vmatrix}$$
