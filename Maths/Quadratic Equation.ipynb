## The Roots
Any equation of the form $ax^2 + bx + c = 0$ where $a \ne 0$ is called a quadratic equation.  

$$ax^2 + bx + c = 0$$
$$x^2 + \frac{b}{a}x + \frac{c}{a} = 0$$
$$(x + \frac{b}{2a})^2 = \frac{b^2}{4a^2} - \frac{c}{a}$$
From here we can see that the two roots are,
$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$
The quantity $b^2 - 4ac$ is called the *discriminant* of the equation.

### Nature of roots
Let $D$ be the discriminant, then
- If $D>0$ then we get two real unequal roots
- If $D<0$ then we get two imaginary roots
- If $D=0$ then we get two equal roots

A quadratic with rational coefficients has rational roots only when $D$ is a perfect square.

**Question:** Let $p$ and $q$ be odd integers, prove that the equation $x^2 + 2px + 2q = 0$ has no rational roots.  
**Answer:** For the above equation to have rational roots, the discriminant should be perfect square:  
$$4p^2 - 8q = {\lambda}^2$$
$$\Rightarrow p^2 - 2q = k^2$$
$$\Rightarrow \frac{p^2 - k^2}{2} = q$$

1) Lets say $k$ is odd, then
$$p^2 - k^2 = (p - k)(p + k)$$
$$\Rightarrow p^2 - k^2 = even \cdot even$$
Thus $\frac{p^2 - k^2}{2}$ is even, contradicting assumption that $q$ is odd.

2) Lets say $k$ is even, then
$$p^2 - k^2 = (p - k)(p + k)$$
$$\Rightarrow p^2 - k^2 = odd \cdot even$$
Thus $\frac{p^2 - k^2}{2}$ is even, contradicting assumption that $q$ is odd.

**Question:** Find $k \in N$, such that the equation $x^2 - kx - 24 = 0$ has integral roots.  
**Answer:**  
$$D = k^2 + 96 = {\lambda}^2$$
$$\Rightarrow {\lambda}^2 - k^2 = 96$$
$$\Rightarrow (\lambda - k)(\lambda + k) = 96$$

Now $(\lambda - k)$ and $(\lambda + k)$ can assume the following values:  
$$(\lambda - k) = 1\ ,\ (\lambda + k) = 96$$
$$(\lambda - k) = 2\ ,\ (\lambda + k) = 48$$
$$(\lambda - k) = 3\ ,\ (\lambda + k) = 32$$
$$(\lambda - k) = 4\ ,\ (\lambda + k) = 24$$
$$(\lambda - k) = 24\ ,\ (\lambda + k) = 4$$
$$(\lambda - k) = 32\ ,\ (\lambda + k) = 3$$
$$(\lambda - k) = 48\ ,\ (\lambda + k) = 2$$
$$(\lambda - k) = 96\ ,\ (\lambda + k) = 1$$

Based on above relations, the possible value of $k$ can be 23, 10, 5 or 2.

### Relation Between Roots and Coefficients
Let $\alpha$ and $\beta$ be the two roots of the equation $ax^2 + bx +c = 0$, then we can write, $ax^2 + bx + c = a(x - \alpha)(x - \beta)$. If we compare this with the equation we get that,
$$\alpha + \beta = -\frac{b}{a}$$
$$\alpha\beta = \frac{c}{a}$$

To derive the above relationship, we:
$$(a{\alpha}^2 + b\alpha + c) - (a{\beta}^2 + b\beta + c) = 0$$
$$\Rightarrow a({\alpha}^2 - {\beta}^2) + b(\alpha - \beta) = 0$$
$$\Rightarrow a(\alpha - \beta)(\alpha + \beta) + b(\alpha - \beta) = 0$$
$$\Rightarrow \alpha + \beta = -\frac{b}{a}$$

To derive the product relationship, we:
$$(a{\alpha}^2 + b\alpha + c) + (a{\beta}^2 + b\beta + c) = 0$$
$$\Rightarrow a({\alpha}^2 + {\beta}^2) + b(\alpha + \beta) + 2c = 0$$
$$\Rightarrow a({\alpha}^2 + {\beta}^2 + 2\alpha\beta - 2\alpha\beta) + b(\alpha + \beta) + 2c = 0$$
$$\Rightarrow a((\alpha + \beta)^2 - 2\alpha\beta) - \frac{b^2}{a} + 2c = 0$$
$$\Rightarrow a(\frac{b^2}{a^2} - 2\alpha\beta) - \frac{b^2}{a} + 2c = 0$$
$$\Rightarrow \alpha\beta = \frac{c}{a}$$

For a cubic equation, the following relations hold,
$$\alpha + \beta + \gamma = -\frac{b}{a}$$
$$\alpha\beta\gamma = -\frac{d}{a}$$
$$\alpha\beta + \beta\gamma + \gamma\alpha = \frac{c}{a}$$

### Symmetric Function of Roots
Let roots of $ax^2 + bx + c = 0$ be $\alpha$ and $\beta$, then
- to get equation having roots $\frac{1}{\alpha}$, $\frac{1}{\beta}$, replace $x$ with $\frac{1}{x}$
- to get equation having roots $\alpha^2$, $\beta^2$, replace $x$ with $\sqrt{x}$
- to get equation having roots $\alpha + k$, $\beta + k$, replace $x$ with $x - k$

**Question**: Let roots of $ax^2 + bx + c = 0$ be $\alpha$ and $\beta$. Find the equation whose roots are $\alpha^2\beta$ and $\beta^2\alpha$.  
**Answer**: Let the desired equation be $Ax^2 + Bx + C =0$. Now the following statements are true
$$\alpha + \beta = - (b/a)$$
$$\alpha\beta = (c/a)$$
Also,
$$\alpha^2\beta + \beta^2\alpha = -(B/A)$$
$$\Rightarrow \alpha\beta(\alpha + \beta) = -(B/A)$$
$$\Rightarrow \frac{bc}{a^2} = (B/A)$$
And,
$$\alpha^2\beta\beta^2\alpha = (C/A)$$
$$\Rightarrow \alpha^3\beta^3 = (C/A)$$
$$\Rightarrow \frac{c^3}{a^3} = (C/A)$$
Therefore the desired equation is $a^3x + abcx + c^3 = 0$

**Question:** Consider the equation $2x^2 -kx +8 \ge 0\ \forall x \in R$, find possible values of $k$.  
**Answer:** The discriminant $D$ must be less than or equal to 0.
$$k^2 - 64 \le 0$$
$$\Rightarrow k^2 \le 8^2$$
Thus
$$k \in [-8, 8]$$

## Common Roots
Consider the equations $ax^2 + bx + c =0$ and $dx^2 + ex + f = 0$. For these two to have both the roots common,
$$\frac{b}{a} = \frac{e}{d}$$
$$\frac{c}{a} = \frac{f}{d}$$
Which leads to,
$$\frac{d}{a} = \frac{e}{b} = \frac{f}{c}$$
$$\Rightarrow \frac{a}{d} = \frac{b}{e} = \frac{c}{f}$$

For atleast one common root,
$$
\begin{vmatrix}
c & a\\
f & d\\
\end{vmatrix}^2 = 
\begin{vmatrix}
b & c\\
e & f\\
\end{vmatrix}\cdot
\begin{vmatrix}
a & b\\
d & e\\
\end{vmatrix}
