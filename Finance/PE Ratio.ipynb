## Overview
P/E Ratio (Price to Earnings Ratio) is the ratio of company's share price to its earnings per share. It is not available for companies that are newly listed and have not reported earnings or are in loss (-ve profit).

$$P/E\ Ratio = \frac{Market\ value\ per\ share}{Earnings\ per\ share}$$

Earning per share is how much a profit company generates per per outstanding share. The earnings of last 4 quarters are used (trailing 12 months). So, on May 26 2026, we get earnings of Q2 2025, Q3 2025, Q4 2025 and Q1 2026.

$$EPS = \frac{Net\ Income}{Total\ Outstanding\ Shares}$$

As an example:  
**AAPL:** on May 22 2026,

$$EPS = \frac{\$83\ billion}{10\ billion} = \$8.3$$
$$P/E\ Ratio = \frac{\$302.25}{\$8.3} = 36.46$$

**TSLA:** on May 22 2026,

$$EPS = \frac{\$12\ billion}{10\ billion} = \$1.2$$
$$P/E\ Ratio = \frac{\$426.01}{\$1.2} = 355.01$$

## Trailing vs Forward P/E Ratio
The above calculation is for trailing P/E ratio based on factual data. Sometimes, the company's projected profit can be used instead which can be then used to calculate forward P/E ratio. Leets say analysts estimate Apple will earn $9.50 per share in the next 12 months:

$$Forward\ P/E = \frac{\$302.25}{\$9.50} = 31.8$$

## Interpretation
P/E ratio tells us how much investors are willing to pay for each unit of earnings. Generally:

| P/E Ratio | Interpretation          | Example                                              |
|-----------|-------------------------|------------------------------------------------------|
| Below 15  | Undervalued/Cheap Stock | Stock may be a bargain, or company may have problems |
| 15-25     | Fairly Valued           | Stock is reasonably priced relative to earnings      |
| 25-50     | Expensive/Growth        | Stock Market expects strong future growth            |
| Above 50  | Very Expensive          | High growth expectations or speculation              |

P/E ratios are not comparable across sectors. As an example, tech industry average: ~25-35, whereas utility industry average is ~12-18.
