## Object Hierarchy
- **Figure** : outermost container for matplotlib graphics
    - **Axes** : a figure can contain multiple axes objects, it essentially represents one plot
        - **Ticks**
        - **Legends**
        - and many more
  
![Object Hierarchy](https://files.realpython.com/media/anatomy.7d033ebbfbc8.png)

```python
# Hierarchy explained
import matplotlib.pyplot as plt
import numpy as np
fig, axes = plt.subplots()
y_axis = fig.axes[0].yaxis
major_ticks = y_axis.get_major_ticks()[0]
major_tick_label = major_ticks.label
```

## Pyplot Internals
Matplotlib implicitly tracks the current axes object. Calling `plt.plot()` calls the plot method of the current axes. Pyplot is basically a bunch of wrapper functions around matplotlib's object oriented interface. For example, `plt.title()` calls the current axes' `set_title()` method. Most of the plt's functions do the same. They first get reference to the current axes and then call its relevant method.

## Subplots
Calling `plt.subplots()` is equivalent to calling `plt.subplots(nrows=1, ncols=1)`

```python
rng = np.arange(50)
yrs = rng + 1950
rnd = np.random.randint(0, 10, (3, 50))

fig, axs = plt.subplots(figsize=(5,3))
axs.stackplot(yrs, rnd + rng, labels=['Asia', 'Africa', 'Europe'])
axs.set_title('Debt Growth')
axs.set_xlabel('year')
axs.set_ylabel('percentage')
axs.set_xlim(xmin=yrs[0], xmax=yrs[-1])

fig.tight_layout()
fig.show()
```

```python
### Multiple Axes
fig2, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8,4))

x = np.random.randint(1,11,50)
y = x + np.random.randint(1,5,x.size)
data = np.column_stack((x,y))

ax1.scatter(x, y, marker='o', edgecolor='b')
ax1.set_title('Scatter Plot')

ax2.hist(data, bins=np.arange(data.min(), data.max()))
ax2.set_title('Histogram')
ax2.yaxis.tick_right()

fig2.show()
```
