## Replication
Redis uses a *master-replica* configuration. Replicas are by default read only and would reject any writes sent to it. The master receives all writes and can asynchronously (or synchronously) replicate it to all replicas. The replicas can be then used for reads.

![Redis Replication Mechanism](./images/redis_replica_basic.png)

Replication allows for:
- high availability, if master goes down, a replica can take over
- distributing read load across multiple nodes
- replication in lieu of persistence

Since network partitioning is unavoidable, there can be 3 scenarios:
- No network patition: master streams all commands to replicas. Replicas become eventually consistent with the master.
- Network temporarily partitioned: when the connection is reestablised, replica may be missing some updates. It will therefore attempt *partial synchronization* with the master.
- Partial synchronization fails: replica attempts *full synchronization*.
