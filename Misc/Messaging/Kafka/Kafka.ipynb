## Kafka
Kafka is a distributed commit log which stores data *durably*, in order and reads are *deterministic*. Data can be distributed acrss multiple *brokers* providing protections against failures, as well as significant opportunities for scaling performance.

## Components
### Message
Is the unit of data in Kafka. It doesn't have a special format, it is just an array of bytes.  
<img src="images/kafka_message.png" />

The payload can be in any format JSON, XML, Avro, etc. Messages are written in batches by combining messages belonging to the same topic and partition. This improves throughput on the expense of latency.

### Topic
While message is akin to a row, a Kafka topic is like a table. It is a way to categorize messages. 

A topic is broken down into multiple **partition**s. Message ordering is guranteed across partition (not across topic). Partitions are also the way that Kafka provides redundancy and scalability. Each partition can be hosted on a different server, which means that a single topic can be scaled horizontally across multiple servers.

Usually the number of partitions for a topic is chosen to be equal to the number of brokers in the cluster. There are several factors to consider while determining the correct number of partitions:
- Expected throughput on the topic. 100KB or 1GB per second?
- Maximum throughput expected from a single partition? At any time, we can have a maximum of one consumer reading from a partition. So in a way the throughput would be limited by the slowest consumer
- If we are sending messages to partitions based on keys, adding partition later can be very challenging.
- Consider number of partitions being placed on each broker in terms of diskspace and network bandwidth.


### Broker
Broker is a single Kafka server. Broker has a number of responsibilities including:
- processing write and read request from producers and consumers
- hosts partitions and clears data after set retention policy
- tracks offsets for consumer groups

Collection of brokers work together forming a *cluster*. Brokers are assigned different roles:
<div style="display: inline-block">

| Role                       | Responsibility                                                                             |
|----------------------------|--------------------------------------------------------------------------------------------|
| Controller (1 per Cluster) | Managing cluster state and administrative tasks.                                           |
| Leader (1 per Partition)   | Handling all read/write requests for a specific data set.                                  |
| Follower (per Partition)   | Replicating data from the leader to provide safety.                                        |
</div>

All producers must connect to the leader in order to publish messages, but consumers may fetch from either the leader or one of the followers.
