## Hypervisor Virtualization
Provides the ability to run multiple operating systems simultaneously on a single physical hardware. This involves creation of *virtual machines* by abstracting away the hardware and divides it among the VMs. Operating system is installed on individual VMs each thinking it has its own hardware.

There are two types of hypervisors:
- **bare metal:** hypervisor runs directly on top of the hardware, no OS underneath. Examples include Microsoft Hyper-V, Amazon Nitro, Xen, etc.
- **hosted:** hypervisor runs on top of an OS like any other program. Examples include VirtualBox, VMWare WorkStation, etc.

## Container Virtualization
Instead of virtualizing hardware, container virtualization virtualizes the operation system. Container share the operating system kernel and just isolate the processes running on it. Applications, libraries and other dependencies are separated out into each container.

Instead of using hypervisors, container virtualization use facilities provided by the operating system. Taking the example of Linux, containerization is made possible using two features:
- **namespaces:** partition kernel resources such that one set of processes see one set of resources while another set of processes see a different set of resources. Each container gets its own process tree (process namespace), network (network namespace), filesystem (mount namespace), users, hostnames, etc.
- **cgroups:** provide resource limiting by sharing CPU, memory, network among all containers.

Other OS have their own technology:
- Solaris provides Zones
- BSD provides Jails
- Windows Containers
